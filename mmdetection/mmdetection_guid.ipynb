````python
from mmengine.utils import get_git_hash
from mmengine.utils.dl_utils import collect_env as collect_base_env

import mmdet

def collect_env():
    """Collect the information of the running environments."""
    env_info = collect_base_env()
    env_info['MMDetection'] = f'{mmdet.__version__}+{get_git_hash()[:7]}'
    return env_info


if __name__ == '__main__':
    for name, val in collect_env().items():
        print(f'{name}: {val}')
````

### [중요] 생성된 data 폴더 아래에 데이터 zip 업로드
SoClean -> mmdetection -> data

#### 압축해제 코드

````python
import zipfile
import os

# 업로드한 파일 이름 입력
filename = "car-interior-v2.v3-mmdetection.coco-mmdetection" # 파일 이름에 맞게 수정

# 업로드된 파일 경로
uploaded_file_path = f"./data/{filename}.zip"  

# 압축 해제할 디렉토리
extracted_dir = f"./data/{filename}"

# 업로드된 파일 압축 해제
with zipfile.ZipFile(uploaded_file_path, 'r') as zip_ref:
    zip_ref.extractall(extracted_dir)

# 압축 해제된 파일 목록 출력
extracted_files = os.listdir(extracted_dir)
print("Extracted files:", extracted_files)
````

````python
import os
import shutil

# 변경할 파일명 및 새로운 파일명 설정
file_mapping = {
    'train/_annotations.coco.json': 'train/_annotations.train.json',
    'test/_annotations.coco.json': 'test/_annotations.test.json',
    'valid/_annotations.coco.json': 'valid/_annotations.valid.json',
}

# 파일 이동 및 이름 변경
for old_name, new_name in file_mapping.items():
    old_path = os.path.join(extracted_dir, old_name)
    new_path = os.path.join(extracted_dir, new_name)
    os.rename(old_path, new_path)
    shutil.move(new_path, extracted_dir)
          
print("Files renamed and moved successfully.")
````
